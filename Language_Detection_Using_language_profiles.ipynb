# Reading the Language Profiles

```python
import os
import json
directory = './profiles'
lis_dic = []
cnt=0
for filename in os.listdir(directory):
    f = open('./profiles/'+filename)
    line = f.readline()
    _dict = json.loads(line)
    _dict = _dict['freq']
    _dict['language_code']=filename
    lis_dic.append(_dict)
_dict1 = lis_dic[0]
print(_dict1['language_code'])
for key,value in _dict1.items():
    print(key)
    print(type(value))
```

```python
cnt=0
for _dict in lis_dic:
    total_freq=0
    for key,value in _dict.items():
        if key!= 'language_code':
            total_freq = total_freq+value
    if cnt==0:
        print('total_freq: '+str(total_freq))
    for key in _dict.keys():
        if key!= 'language_code':
            if cnt==0:
                print('value2: '+str(_dict[key])+' total_freq: '+str(total_freq))
            _dict[key] = (_dict[key]/total_freq)
    cnt += 1
```

```python
#utility function
def multiply_list_dict(num):
    for _dict in lis_dic:
        for key in _dict.keys():
            if key!= 'language_code':
                _dict[key] = _dict[key]*num
    return
```

# Language Detector (implementation of Bayes Rule on Tri-grams)

```python
import re
import numpy as np

def find_language(input_text, lis_dict_p):
    input_text = re.sub(r'[^a-zA-Z0-9\s]', ' ', input_text)
    n = len(input_text)
    #finding the probability
    multiply_list_dict(n)
    num_lang = len(lis_dict_p)
    prob = np.zeros(num_lang)
    for i in range(num_lang):
        ck_dict = lis_dict_p[i]
        temp = len(lis_dict_p[0])-1  #1/temp will be default probability value
        p_ck_x = 1/50
        for j in range(0,n):
            x = min(n,j+3)
            n_gram = input_text[j:x]
            if n_gram in ck_dict.keys():
                p_ck_x = p_ck_x * ck_dict[n_gram]
            else:
                p_ck_x = p_ck_x * 1/50
        prob[i]=p_ck_x
    max_ind = np.argmax(prob)
    return lis_dict_p[max_ind]['language_code']
```
